# Notebook 04: Feature Engineering

## Objective

This notebook prepares task-specific datasets for machine learning.

The preprocessing performed here includes:

- Removing identifier variables
- Selecting target variables
- Creating feature matrices for each prediction task
- Saving task-specific datasets

Categorical encoding and feature scaling will be performed inside machine learning pipelines during model training to prevent data leakage.

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
import numpy as np
import pandas as pd
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
from src.config import PROCESSED_DATA_DIR
```

```python
df = pd.read_csv(
    PROCESSED_DATA_DIR / "cleaned_cycle_data.csv"
)

print(df.shape)

df.head()
```

## Remove Identifier Columns

Identifiers do not contribute to prediction and may cause overfitting.

```python
identifier_columns = [
    "ClientID",
    "CycleNumber"
]

identifier_columns = [
    col for col in identifier_columns
    if col in df.columns
]

df = df.drop(columns=identifier_columns)

print(df.shape)
```

```python
print(df.dtypes.value_counts())
```

## Encode Variables

Machine learning models require numerical input.

```python
categorical_features = df.select_dtypes(include="object").columns.tolist()

numerical_features = df.select_dtypes(exclude="object").columns.tolist()

print("Categorical:", len(categorical_features))
print("Numerical :", len(numerical_features))
```

# Remove Potential Leakage Features

```python
leakage_features = [

    "FirstDayofHigh",

    "TotalNumberofHighDays",

    "TotalHighPostPeak",

    "TotalNumberofPeakDays",

    "TotalDaysofFertility",

    "TotalFertilityFormula"

]

df = df.drop(columns=leakage_features)

df.shape
```

# Dataset 1

## Predict Estimated Day of Ovulation

```python
target = "EstimatedDayofOvulation"

X_ovulation = df.drop(columns=[target])

y_ovulation = df[target]

print(X_ovulation.shape)
```

# Dataset 2

## Predict Cycle Length

```python
target = "LengthofCycle"

X_cycle = df.drop(columns=[target])

y_cycle = df[target]

print(X_cycle.shape)
```

# Dataset 3

## Predict Cycle With Peak or Not

```python
target = "CycleWithPeakorNot"

X_peak = df.drop(columns=[target])

y_peak = df[target]

print(X_peak.shape)
```

```python
print(X_ovulation.columns.tolist())
```

# Feature Summary

```python
summary = pd.DataFrame({

    "Dataset":[
        "Ovulation",
        "Cycle Length",
        "Peak Detection"
    ],

    "Features":[
        X_ovulation.shape[1],
        X_cycle.shape[1],
        X_peak.shape[1]
    ],

    "Samples":[
        X_ovulation.shape[0],
        X_cycle.shape[0],
        X_peak.shape[0]
    ]

})

summary
```

```python
summary.to_csv(
    PROCESSED_DATA_DIR / "dataset_summary.csv",
    index=False
)
```

```python
X_ovulation.to_csv(
    PROCESSED_DATA_DIR / "X_ovulation.csv",
    index=False
)

y_ovulation.to_csv(
    PROCESSED_DATA_DIR / "y_ovulation.csv",
    index=False
)

X_cycle.to_csv(
    PROCESSED_DATA_DIR / "X_cycle.csv",
    index=False
)

y_cycle.to_csv(
    PROCESSED_DATA_DIR / "y_cycle.csv",
    index=False
)

X_peak.to_csv(
    PROCESSED_DATA_DIR / "X_peak.csv",
    index=False
)

y_peak.to_csv(
    PROCESSED_DATA_DIR / "y_peak.csv",
    index=False
)

print("Datasets saved successfully.")
```
