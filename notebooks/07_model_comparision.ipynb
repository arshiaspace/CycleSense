```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("ggplot")
```

```python
import pandas as pd

from src.config import REPORT_DIR

regression_results = pd.read_csv(
    REPORT_DIR / "regression_results.csv"
)

classification_results = pd.read_csv(
    REPORT_DIR / "classification_results.csv"
)

regression_results
```

```python
regression_results
```

```python
classification_results
```

```python
plt.figure(figsize=(8,5))

sns.barplot(
    data=regression_results,
    x="RMSE",
    y="Model"
)

plt.title("Regression Model Comparison")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.barplot(
    data=classification_results,
    x="F1",
    y="Model"
)

plt.title("Classification Model Comparison")

plt.show()
```

```python
best_regression = regression_results.loc[
    regression_results["RMSE"].idxmin()
]

best_regression
```

```python
best_classifier = classification_results.loc[
    classification_results["F1"].idxmax()
]

best_classifier
```

# Discussion

## Regression

The regression models were evaluated using MAE, RMSE, and R².

## Classification

Classification models were evaluated using Accuracy, Precision, Recall, F1-score, and ROC-AUC.
