## Notebook 02: Data Cleaning

### Objective

The objective of this notebook is to clean the raw menstrual cycle dataset by identifying missing values, removing duplicate records, handling inconsistent data, eliminating irrelevant features, and preparing a clean dataset for exploratory data analysis and machine learning.

```python
df_clean.replace(r'^\s*$', np.nan, regex=True, inplace=True)
```

```python
numeric_columns = [
    "MeanCycleLength",
    "EstimatedDayofOvulation",
    "LengthofLutealPhase",
    "FirstDayofHigh",
    "TotalNumberofHighDays",
    "TotalHighPostPeak",
    "TotalNumberofPeakDays",
    "TotalDaysofFertility",
    "TotalFertilityFormula",
    "LengthofMenses",
    "MeanMensesLength",
    "MensesScoreDayOne",
    "MensesScoreDayTwo",
    "MensesScoreDayThree",
    "MensesScoreDayFour",
    "MensesScoreDayFive",
    "MensesScoreDaySix",
    "MensesScoreDaySeven",
    "MensesScoreDayEight",
    "MensesScoreDayNine",
    "MensesScoreDayTen",
    "MensesScoreDay11",
    "MensesScoreDay12",
    "MensesScoreDay13",
    "MensesScoreDay14",
    "MensesScoreDay15",
    "TotalMensesScore",
    "MeanBleedingIntensity",
    "NumberofDaysofIntercourse",
    "IntercourseInFertileWindow",
    "UnusualBleeding",
    "PhasesBleeding",
    "IntercourseDuringUnusBleed",
    "Age",
    "AgeM",
    "Maristatus",
    "MaristatusM",
    "Yearsmarried",
    "Religion",
    "ReligionM",
    "Ethnicity",
    "EthnicityM",
    "Schoolyears",
    "SchoolyearsM",
    "OccupationM",
    "IncomeM",
    "Height",
    "Weight",
    "Reprocate",
    "Numberpreg",
    "Livingkids",
    "Miscarriages",
    "Abortions",
    "Medvits",
    "LivingkidsM",
    "Boys",
    "Girls",
    "MedvitsM",
    "Breastfeeding",
    "Method",
    "Prevmethod",
    "Whychart",
    "Nextpreg",
    "NextpregM",
    "Spousesame",
    "SpousesameM",
    "Timeattemptpreg",
    "BMI"
]

for col in numeric_columns:
    if col in df_clean.columns:
        df_clean[col] = pd.to_numeric(df_clean[col], errors="coerce")
```

```python
df_clean.dtypes.value_counts()
```

```python
numeric_cols = df_clean.select_dtypes(include=np.number).columns
categorical_cols = df_clean.select_dtypes(exclude=np.number).columns

for col in numeric_cols:
    df_clean[col] = df_clean[col].fillna(df_clean[col].median())

for col in categorical_cols:
    mode = df_clean[col].mode()
    if not mode.empty:
        df_clean[col] = df_clean[col].fillna(mode[0])
```

```python
print("Remaining Missing Values:", df_clean.isnull().sum().sum())
```

```python
from src.config import PROCESSED_DATA_DIR

# Ensure directory exists
PROCESSED_DATA_DIR.mkdir(parents=True, exist_ok=True)

cleaned_data_path = PROCESSED_DATA_DIR / "cleaned_cycle_data.csv"

df_clean.to_csv(
    cleaned_data_path,
    index=False
)

print("="*50)
print("Cleaned Dataset Saved Successfully")
print("="*50)
print(f"Location : {cleaned_data_path}")
print(f"Shape    : {df_clean.shape}")
```

```python
summary = pd.DataFrame({
    "Metric": [
        "Rows",
        "Columns",
        "Remaining Missing Values",
        "Duplicate Rows"
    ],
    "Value": [
        df_clean.shape[0],
        df_clean.shape[1],
        df_clean.isnull().sum().sum(),
        df_clean.duplicated().sum()
    ]
})

summary
```

```python
from src.config import REPORT_DIR

# Ensure reports directory exists
REPORT_DIR.mkdir(parents=True, exist_ok=True)

summary_path = REPORT_DIR / "cleaning_summary.csv"

summary.to_csv(
    summary_path,
    index=False
)

print(f"Cleaning summary saved to:\n{summary_path}")
```

The following preprocessing steps were completed successfully:

- Converted numeric-like object columns into numerical data types
- Imputed missing numerical values using the median
- Imputed missing categorical values using the mode
- Verified the absence of remaining missing values
- Saved the cleaned dataset and summary reports

The cleaned dataset is now ready for feature understanding and exploratory data analysis.
